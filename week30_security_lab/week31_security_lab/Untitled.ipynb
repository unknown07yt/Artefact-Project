```python
def perm_to_octal(perm):
    if len(perm) != 9:
        return "Invalid permission string"

    mapping = {'r':4, 'w':2, 'x':1, '-':0}
    result = ""

    for i in range(0, 9, 3):
        value = 0
        value += mapping.get(perm[i],0)
        value += mapping.get(perm[i+1],0)
        value += mapping.get(perm[i+2],0)
        result += str(value)

    return result
```

```python
perm_to_octal("rwxr-xr--")
```

```python
perm = input("Enter permission string (e.g. rwxr-xr--): ")
print("Octal:", perm_to_octal(perm))
```
